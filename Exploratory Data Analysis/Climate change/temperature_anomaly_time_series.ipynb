```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
df = (
    pd.read_csv("global_latestdata.csv")
    .set_index('Year', verify_integrity=True)
)
df_north = (
    pd.read_csv("northamerica.csv")
    .set_index('Year', verify_integrity=True)
)
df_south = (
    pd.read_csv("southamerica.csv")
    .set_index('Year', verify_integrity=True)
)
df_asia = (
    pd.read_csv("asia.csv")
    .set_index('Year', verify_integrity=True)
)
df_africa = (
    pd.read_csv("africa.csv")
    .set_index('Year', verify_integrity=True)
)
df_eur = (
    pd.read_csv("europe.csv")
    .set_index('Year', verify_integrity=True)
)
df_ocean = (
    pd.read_csv("oceania.csv")
    .set_index('Year', verify_integrity=True)
)
```

```python
fig, ax = plt.subplots(figsize=(10,6))

# Plot temperature data with specific colour and line thickness
ax.plot(df, color='#1C7C54', linewidth=2.5,label='annual_mean')

# Set axis labels and graph title
ax.set(xlabel='Time (years)', ylabel='Temperature Anomaly (deg. Celsius)',
       title='Global Temperature Anomalies')

# Enable grid
ax.grid()
#df.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
plt.legend(loc="upper left")
```

```python
fig, ((ax1,ax2,ax3), (ax4, ax5,ax6)) = plt.subplots(2, 3, figsize=(20,16))
#fig, ax1 = plt.subplots(figsize=(10,6))

#asia
ax1.plot(df_asia, color='#ff7f0e', linewidth=2.5,label='annual_mean')
ax1.set(xlabel='Time (years)', ylabel='Temperature Anomaly (deg. Celsius)',
       title='Asia')
ax1.grid()
ax1.set_xlabel('Year');

#africa
ax2.plot(df_africa, color='#1f77b4', linewidth=2.5,label='annual_mean')
ax2.set(xlabel='Time (years)', ylabel='Temperature Anomaly (deg. Celsius)',
       title='Africa')
ax2.grid()
ax2.set_xlabel('Year');

#europe
ax3.plot(df_eur, color='#bcbd22', linewidth=2.5,label='annual_mean')
ax3.set(xlabel='Time (years)', ylabel='Temperature Anomaly (deg. Celsius)',
       title='Europe')
ax3.grid()
ax3.set_xlabel('Year', fontsize=20);

#north america
ax4.plot(df_north, color='#17becf', linewidth=2.5,label='annual_mean')
ax4.set(xlabel='Time (years)', ylabel='Temperature Anomaly (deg. Celsius)',
       title='North America')
ax4.grid()
ax4.set_xlabel('Year', fontsize=20);

#south america
ax5.plot(df_south, color='#9467bd', linewidth=2.5,label='annual_mean')
ax5.set(xlabel='Time (years)', ylabel='Temperature Anomaly (deg. Celsius)',
       title='South America')
ax5.grid()
ax5.set_xlabel('Year',fontsize=20);

#oceania
ax6.plot(df_ocean, color='#7f7f7f', linewidth=2.5,label='annual_mean')
ax6.set(xlabel='Time (years)', ylabel='Temperature Anomaly (deg. Celsius)',
       title='Oceania')
ax6.grid()
ax6.set_xlabel('Year', fontsize=20);

ax1.set_ylabel('Temperature Anomaly (deg. Celsius)',fontsize=15);
ax2.set_ylabel('Temperature Anomaly (deg. Celsius)',fontsize=15);
ax3.set_ylabel('Temperature Anomaly (deg. Celsius)',fontsize=15);
ax4.set_ylabel('Temperature Anomaly (deg. Celsius)',fontsize=15);
ax5.set_ylabel('Temperature Anomaly (deg. Celsius)',fontsize=15);
ax6.set_ylabel('Temperature Anomaly (deg. Celsius)',fontsize=15);

ax1.set_title('Asia', fontsize=20);
ax2.set_title('Africa', fontsize=20);
ax3.set_title('Europe', fontsize=20);
ax4.set_title('North America', fontsize=20);
ax5.set_title('South America', fontsize=20);
ax6.set_title('Oceania', fontsize=20);

ax1.legend(loc="upper left")
ax2.legend(loc="upper left")
ax3.legend(loc="upper left")
ax4.legend(loc="upper left")
ax5.legend(loc="upper left")
ax6.legend(loc="upper left")

fig.subplots_adjust(hspace=0.3)
plt.show()
fig.savefig('regions_temp.png')
```
